```python
"""
Function
--------
parse_candidates

Given the top two contendors that meet the following criteria:

1) It's a list of dictionaries from the Huffington Post chart API which has a key for "estimates"
2) It is during a year in which one of the top two condentors is always a Republican
3) It is the fourth homework assighment of the fall 2014 CS-109 Data Science class.

returns a dictionary of the full name of the republican, the full name of the 
Democrat or "other" (Independant, Libertarian, Green, etc), and whether either is an 
incumbent.  This also includes a test called "actual" so you can see if the conversion
of other to Democrat is correct.

Inputs
------
contenders : List
    A list of two dictionaries of the top two contenders in a race that meets
    the specifications outlined above.

Returns
-------
dict : Dict
    A dictionary with the following structure where R is the fullname of the
    Republican candidate, D is the full name of the Democrat or "other", and the
    Incumbent is the party affilaiton of the incument or NA.  "Actual" is a test
    with which you can tell if the other to Democrat convestion was successful.
    
    {"R" : R, "D" : D, "Incumbent" : incumbent, "Actual" : actual}
    
Example: 
>>> choices = [{u'choice': u'Hagan',
  u'first_name': u'Inde',
  u'incumbent': True,
  u'last_name': u'pendant',
  u'lead_confidence': 94.6,
  u'party': u'Ind',
  u'value': 42.3},
 {u'choice': u'Tillis',
  u'first_name': u'Repub',
  u'incumbent': False,
  u'last_name': u'Lican',
  u'lead_confidence': 5.4,
  u'party': u'Rep',
  u'value': 40.1},
 {u'choice': u'Gennaro',
  u'first_name': u'Demo',
  u'incumbent': False,
  u'last_name': u'Crat',
  u'lead_confidence': 1.4,
  u'party': u'Dem',
  u'value': 40.1}]
>>> parse_candidates(choices[0:2])
>>> {'Actual': u'Ind', 'D': u'Inde pendant', 'Incumbent': 'D', 'R': u'Repub Lican'}
"""

def parse_candidates(contenders):

    incumbent = unicode("NA")
    actual = unicode("none")

    for choice in contenders:

        # SORT OUT WHO IS A "R" (REPULBICAN) OR "D" (DEMOCRATE OR NEITHER)

        # republican?
        if choice["party"] == "Rep":
            R = choice["first_name"] + " " + choice["last_name"]
            # also incumbent?
            if choice["incumbent"] == True:
                incumbent = unicode("R")
                actual = choice["party"] # validation
        # democrat or niether?
        else:
            D = choice["first_name"] + " " + choice["last_name"]
            # also incumbent?
            if choice["incumbent"] == True:
                incumbent = unicode("D")
                actual = choice["party"] # validation
                
    return {"R" : R, "D" : D, "Incumbent" : incumbent, "Actual" : actual} 
```

```python
# PART 1: create a list of each state along with their last names, we'll need the state for annotating
r_state_last = pd.DataFrame([[state, name.split()[-1:][0]] for state, name in zip(candidates["state"], candidates["R"])])
r_state_last.columns=["state", "lastname"]
r_state_last.head()
```

```python
# PART 2 SUBNOTE: I realize that I could "condense" everything below by not always extracting the column of the dataframe into
# it's own variable.  It's easier for me to see it this way since I'm still trying to become familiar with the formulas 
# themselves
```

```python
# PART 2: compute the standard deviations of the observed poll results for each race. 
tvo = []

# get the last names of republicans so we can get the column with the P's
# for every name in candidates["R"], split it, take the last element in the returned list, then take the string
r_lastnames = [name.split()[-1:][0] for name in candidates["R"]]

for race in polls:
#     print race
   
    # parse last names of candidates 1 and 2 from the key, see note 1 for explaination
    words = race.split()
#     print words
    c1, c2 = words[-1], words[-3]
#     print c1, c2
    
    # test to see which candidate is the republican
    republican = c1 if c1 in r_lastnames else c2
#     print republican

    # go through poll and gather all the p's. then dived by 100
    P = polls[race][republican].div(100)
#     print P

    # then N, do it for the same things, a vector N.
    N = polls[race]["Number of Observations"]
#     print N

    # M is just the length of the N. 
    M = len(N) #5
#     print M

    # calculate the variance of each poll.  
    # using the mean of P and the mean of 1/N's, use P(1-P)*1/N
    # using the mean of the 1/n's
    
    VAR = ((P.mean()*(1- P.mean()))*(1/N).mean())
    
    # nothing more than the square root of the variance
    THSTD = np.sqrt(VAR)
    
    # observed standard deviation
    OBSTD = np.std(P)
    
    # i know, in efficant, but I didn't want to add a sepperate index count
    tvo.append([republican, M, VAR, THSTD, OBSTD]) 

theory_obs = pd.DataFrame(tvo, columns=["R", "POLLS","VAR", "THSTD", "OBSTD"])

theory_obs = pd.merge(left=r_state_last, right=theory_obs, left_on='lastname', right_on='R')

theory_obs.head()

# 1) split up the key into individual words.  when the string is split, the lists are not always the same size
# because some states have two words in their name (e.g. new jersey), but the names will always be the last, and third
# to last of that list.
```

More precisely, it states that as n gets larger, the distribution of the difference between the sample average Sn and its limit µ, when multiplied by the factor √n (that is √n(Sn − µ)), approximates the normal distribution with mean 0 and variance σ2.
