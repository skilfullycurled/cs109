```python
"""
Function
--------
diffsByPlstr

Given a pandas dataframe with polls of an election race that inclues differences of points share of votes,
returns a list of the collected differenecs for each unique pollster.  That is, if there are pollsters with 
more than one poll, all of the differences of the indicidual polls will be put together.  There won't be a
repeition.

Inputs
------
race : pd.DataFrame
    A pandas dataframe of a state senate race.

Returns
-------
diffs : list
    A list with the differences of each poll grouped by pollster.
    
Example: 
# in this case the distrubtion is [653 661 658 656 648]
>>> alaskadiffs = diffsByPlstr(alaska)
>>> alaskadiffs[0][:5]
15    3
28    1
43   -1
51   -1
80    3
"""

def diffsByPlstr(race):
    
    # place to store the differences
    diffs = []
    
    # get all the unique pollsters
    pollsters = list(set(race["Pollster"]))
    
    # grab all the polls from that pollster
    for pollster in pollsters:
        
        # grab all the polls from that pollster
        polls = race[race["Pollster"] == pollster]

        # append their differences to the list of differences
        diffs.append(polls["Diff"])

    # send 'em back out
    return diffs
```

```python
"""
Function
--------
getMeans

Given a list of the differences of polls by the same pollster, returns
a list of means for each poll.

Inputs
------
plstrdiffs : list
    A list of differnces

Returns
-------
means : list
    A list with means of each poll in the lsit of differences
    
Example: 
>>> means = getMeans(alaskadiffs)
>>> means[0][:5]
[0.0,
 3.5,
 2.0,
 2.25,
 5.75,
"""

def getMeans(plstrdiffs):
    
    # place to store the differences
    means = []
    
    for diffs in plstrdiffs:
        means.append(diffs.mean())
        
    return means
```

```python
"""
Function
--------
pickwinner

Given two candidates of a race, mean and standard error, returns a the name of the
winner, or too close to call.

Inputs
------
c1 : str
    Name of first candidate
    
c2: str
    Name of second candidate

mn: float
    mean of averages of all available polls
    
se: float
    standard error

Returns
-------
winner : str
    the name of the winner, or too close to call.
    
Example: 
>>> pickwinner("Oscar", "Dorothy", 0.2, 0.05)
>>> "Oscar"
"""
def pickwinner(c1, c2, mn, se):
    # if with the se factored in both directions the diff is still above 1
    if ((mn - se) >= 0.01) and ((mn + se) >= 0.01):
        # candidate 1 wins
        return c1
    # if with the se factored in both directions the diff is still below 1
    elif ((mn - se) <= -0.01) and ((mn + se) <= -0.01):
        # candidate 2 wins
        return c2
    # if the mean minus se is between -1 and 1
    else:
        # who can say?
        return "Too close to call"
```

```python
for race in senate:
    
    # get the diffs for the particular race
    diffs = diffsByPlstr(senate[race])
    
    # get the means of all the diffs
    means = getMeans(diffs)
    
    # mean of all diff means
    diff_mean = np.mean(means)
    
    # mean of mean
    diff_se = np.std(means)
    
    # choose winner
    winner = pickwinner(senate[race].columns[7], senate[race].columns[8], diff_mean, diff_se)

    # give printout
    print race
    print "Number of Polls", len(senate[race])
    print "Mean:", diff_mean/100 
    print "Standard Error:",diff_se/100
    print "Winner:", winner
    print ""
```
