# Problem 2:  $k$-Nearest Neighbors and Cross Validation 

What is the optimal $k$ for predicting species using $k$-nearest neighbor classification 
on the four features provided by the iris dataset.

In this problem you will get to know the famous iris data set, and use cross validation to select the optimal $k$ for a $k$-nearest neighbor classification. This problem set makes heavy use of the [sklearn](http://scikit-learn.org/stable/) library. In addition to Pandas, it is one of the most useful libraries for data scientists! After completing this homework assignment you will know all the basics to get started with your own machine learning projects in sklearn. 

Future lectures will give further background information on different classifiers and their specific strengths and weaknesses, but when you have the basics for sklearn down, changing the classifier will boil down to exchanging one to two lines of code.

The data set is so popular, that sklearn provides an extra function to load it:

```python
# code based on sklearn example
def visualize_decision_boundary(classifier, xmin, xmax, ymin, 
                                ymax, step_size=0.02, 
                                cmap=cmap_light):
    
    xx, yy = np.meshgrid(np.arange(xmin, xmax, step_size),
                         np.arange(ymin, ymax, step_size))
    colors = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    colors = colors.reshape(xx.shape)
    ax = plt.subplot(111)
    ax.pcolormesh(xx, yy, colors, cmap=cmap)
    return ax

# TruncatedSVD's attribute explained_variance_ratio_ sems to be broken
# http://stackoverflow.com/questions/20563239/truncatedsvd-explained-variance

def evr(model):
    x_proj = model.transform(x)
    explained_variances = np.var(x_proj, axis=0) / np.var(x, axis=0).sum()
    return explained_variances


# also from verena's lection
def accuracy_for_k(k,x,y):
    knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=k, 
                                             weights = 'uniform')
    knn.fit(x,y)
    y_hat = knn.predict(x)
    return sklearn.metrics.accuracy_score(y, y_hat, normalize=True)

def accuracy_for_k_val(k,x,y, random_state=99):
    split_data = sklearn.cross_validation.train_test_split(
                        x,y, test_size=0.33, random_state=random_state)
    x_train, x_val, y_train, y_val = split_data
    knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=k, 
                                             weights = 'uniform')
    knn.fit(x_train,y_train)
    y_hat = knn.predict(x_val)
    return sklearn.metrics.accuracy_score(y_val, y_hat, 
                                          normalize=True)
```

```python
#load the iris data set
iris = sklearn.datasets.load_iris()

# Sepal Length, Sepal Width, Petal Length and Petal Width.
x = iris.data

# type of flower Setosa, Versicolour, and Virginica, as 0, 1, 2.
y = iris.target

print x.shape, y.shape
```

#### Problem 2(c) 

In the lecture we discussed how to use cross validation to estimate the optimal value for $k$ (the number of nearest neighbors to base the classification on). Use ***ten fold cross validation*** to estimate the optimal value for $k$ for the iris data set. 

**Note**: For your convenience sklearn does not only include the [KNN classifier](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html), but also a [grid search function](http://scikit-learn.org/stable/modules/generated/sklearn.grid_search.GridSearchCV.html#sklearn.grid_search.GridSearchCV). The function is called grid search, because if you have to optimize more than one parameter, it is common practice to define a range of possible values for each parameter. An exhaustive search then runs over the complete grid defined by all the possible parameter combinations. This can get very computation heavy, but luckily our KNN classifier only requires tuning of a single parameter for this problem set. 

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report

# what k's do we want?
k = np.arange(1, 50, 2)

# ready the classifier
knn = sklearn.neighbors.KNeighborsClassifier()

# search over all the k's using 10 folds each
parameters = {"n_neighbors": k}

gs = GridSearchCV(knn, param_grid=parameters, cv=10)

# fit the training data
gs.fit(x_train, y_train)

# print the best parameter of K
print "Estimation of the optimal value of K:", gs.best_params_
```

```python
scores = []

# neccessary for forthcoming joke only
sevenyearsago = gs.grid_scores_

for score in sevenyearsago:
    scores.append(score.cv_validation_scores)
    
s = np.array(scores)

sns.tsplot(s.T)
```

```python
# from verena's lecture

acc_values_val = []
acc_values_train = []

k_values = xrange(1,80,2)

for k in k_values:
    acc_values_train.append(accuracy_for_k(k=k,x=x_train,y=y_train))
    acc_values_val.append(accuracy_for_k_val(k=k,x=x_test,y=y_test))
    
plt.plot(k_values, acc_values_train, c='r', label="training accuracy")
plt.plot(k_values, acc_values_val, c='b', label="validation accuracy")
plt.xlabel('Value of k')
plt.ylabel('Accuracy')
plt.legend()

fig_cross_val = plt.gcf()
```

```python
"""
Function
--------
uniqueshuffle

Given a range, returns two np.arrays whose values do not have
the same position in each array.

Inputs
------
r1 : int
    the begining of the range of values

r2 : int
    the exclusive end of the range in values
    
Returns
-------

a, b : two np.arrays whose values do not have the same position.

Example:
>>> c, d = uniqueshuffle(0, 10)
>>> c, d
>>> [9 0 5 7 4 2 6 1 3 8] [1 8 6 3 5 9 0 2 4 7]
"""

# completley unneccessary discursion in recursion 
def uniqueshuffle(r1, r2):
    
    a = np.arange(r1,r2)
    b = np.arange(r1,r2)
    
    np.random.shuffle(a), np.random.shuffle(b)
        
    if False not in np.not_equal(a, b):
        print "here you go!"
        return a, b
    else:
        print "that was fun, let's do it again!"
        return uniqueshuffle(r1, r2)

# fset = feature set
fset1, fset2 = uniqueshuffle(0, 10)

for f1, f2 in zip(fset1, fset2):

    plt.figure(figsize=(5, 5))

    x_sub_2d = x_2d[np.logical_or(y_train == f1, y_train == f2),:]
    y_sub = y_train[np.logical_or(y_train==f1, y_train==f2)]

    plt.scatter(x_sub_2d[:,0], x_sub_2d[:,1], c=y_sub, 
                s = 50, cmap=plt.cm.Set1, vmin=0, vmax=9)
    plt.colorbar()

    plt.xlim(30, 75)
    plt.ylim(-40,40)
    
    plt.title(str(f1) + " & " + str(f2))

    plt.show()
```
