```python
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Advice from cs109 2013, and I think it is good!

#tell pandas to display wide tables as pretty HTML tables
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest
    
join : str
    The type of join you would like, inner or outer.
    Default is inner.

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)

"""

def mergeByYear(year, join="inner"):
    
    # merge via join on countries.Country and income.gdp pc test
    # df countries cols: country, area 
    # df income cols: (gdp pc (country), years (incomes)
    merged = pd.merge(countries, income, left_on="Country", right_on="gdp pc test", how=join)
        
    # get the incomes of the specified year, and ditch the extra country column, gdp pc test
    pieces = [merged["Country"], merged["Region"], merged[year]]
    
    # put the list back into a dataframe
    conc = pd.concat(pieces, axis=1)
    
    # add new column titles, in particular, change year to income since the year is not implied
    conc.columns = ["Country", "Region", "Income"]

    return conc


mergeByYear(1980)

# QUESTION
# I learned through trial and error that axis=1 was neccessary, why?
```

```python
# 1. incomesbyYear()
# I decided to write this to help me describe the entire dataset by region.
# I later came to find that I could also use df.groupby("Region").describe() 
# But this is more compact and only on one hierarchy (is that the word?)

"""
Function
--------
incomesByArea

Returns a merged DataFrame containing the incomes 
and country names of a region for a range of years. 

Parameters
----------
df : pd.DataFrame
    The dataframe of the year in question with columns for 
    "Country", "Region", and "Income".

Returns
-------
a DataFrame
   A pandas DataFrame with columns titled by all of the areas
   and the incomes of each country below

Example
-------
>>> incomesByRegion(dfyear)

"""
# Todo: Better name choice for more generalizability
def incomesByRegion(df):
    
    #list to store pieces of the dataframe
    incomes_by_region = []
    
    #names of future collumns of new dataframe
    regions = df.Region.unique()
    
    # extract incomes of the countries for each area and put in new dataframe
    # cycle through each region
    for region in regions:
        # get the individual area
        area = df[df.Region == region].reset_index() #see note below
        # add the income collumn to the new dataframe
        incomes_by_region.append(area["Income"])
    
    # join the pieces into a new dataframe to be sumarize
    summary = pd.concat(incomes_by_region, axis=1)
    # add the new collumn names
    summary.columns = regions
    
    # return summary dataframe
    return summary

# note: could do it all in one like this but I've chosen not to for legibility
#       incomes.append(df[df.Region == region].reset_index()["Income"])
        
incomesByRegion(mergeByYear(1980)).head()
```

```python
"""
Function
--------
getMeans

Returns a merged DataFrame containing the means 
and years by region for a range of years. 

Parameters
----------
start : int
    The start date of the range of years you would like
    
end : int
    The year after you would like your last date.  So,
    if you want 1989, then the end year is 1990

Returns
-------
a DataFrame
   A pandas DataFrame with columns titled by all of the areas
   and the mean income for each area over the years specified

Example
-------
>>> getMeans(1980, 1990)

"""
def getMeans(start, end):
    
    # list to store the means/areas
    means = []
    # years to concat, also, convenient to iterate over
    years = [year for year in range(start, end)]

    # create a list for each mean/region df
    # loop over the series of years
    for year in years:

        # get the incomes by region for a particular year
        ibr_year = incomesByRegion(mergeByYear(year))
        
        # get the mean row of the description table
        mn_year = ibr_year.describe().ix[1]
        
        means.append(mn_year)
    
    #concat the means into a new dataframe
    total_means = pd.concat(means, axis=1)
    
    #replace column titles with the years
    total_means.columns = years
    
    return total_means

# getMeans(1980, 1983)

means = getMeans(1980,1990)
means.T.head()
```
